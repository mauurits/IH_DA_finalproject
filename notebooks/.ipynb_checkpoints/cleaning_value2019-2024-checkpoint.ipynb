```python
# rename column names

column_mappings = {
    'Perioden': 'region',
    '2019': '2019',
    '2020': '2020',
    '2021': '2021',
    '2022': '2022',
    '2023**': '2023',
    '2024*': '2024'
}

print("\nColumn Mappings Dictionary:")
print(column_mappings)

# Rename the columns in the DataFrame
df1.rename(columns=column_mappings, inplace=True)
```
