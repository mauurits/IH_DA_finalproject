```python
# Strip 'occurred_at' to just the year
df1['year'] = pd.to_datetime(df1['occurred_at']).dt.year

# Create a function to categorize magnitudes
def categorize_magnitude(mag):
    if 1.0 <= mag < 2.0:
        return 1
    elif 2.0 <= mag < 3.0:
        return 2
    elif 3.0 <= mag < 4.0:
        return 3
    else:
        return 0

# Apply the categorization function to create a new column
df1['magnitude_category'] = df1['magnitude'].apply(categorize_magnitude)

# Display the first few rows of the modified DataFrame
print(df1[['year', 'magnitude', 'magnitude_category']].head())

# Optional: Get value counts of magnitude categories
print(df1['magnitude_category'].value_counts())

df1
```
