```python
agg_functions = {
    'occurred_at': 'count',  # Calculate mean for 'magnitude'
    'magnitude': 'max',
    'depth': 'min', # Calculate sum for 'depth'
    'latitude': 'max',
    'longitude': 'max'
}

# Group by 'city' and aggregate other columns
df_grouped = df1.groupby('city').agg(agg_functions)
df_sorted = df_grouped.sort_values(by='occurred_at', ascending=False)
top10 = df_sorted.iloc[0:20]
top10
# top10.to_csv('regions_by_numberoccurrences_top20.csv', index=False)
```
